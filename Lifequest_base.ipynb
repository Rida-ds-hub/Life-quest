```python
import mysql.connector
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import mysql.connector

try:
    # Connect to MySQL database
    connection = mysql.connector.connect(
        user='root',
        password='******',
        host='localhost',
        database='lifequest',
        ssl_disabled=True
    )
    print("Connected to MySQL database successfully!")
except mysql.connector.Error as err:
    print("Error: Failed connecting to MySQL database.")
    print(err)
```

```python
# Create a cursor object
cursor = connection.cursor()
```

```python
#Simple queries 
# 1. Retrieve all academic types
cursor.execute("SELECT Academic_Type FROM ACADEMIC")
academic_types = cursor.fetchall()
print("Academic Types:")
for academic_type in academic_types:
    print(academic_type[0])
```

```python
# 2. Retrieve all achievements with their descriptions and dates achieved
cursor.execute("SELECT Description, Date_Achieved FROM ACHIEVEMENT")
achievements = cursor.fetchall()
print("\nAchievements:")
for achievement in achievements:
    print("Description:", achievement[0])
    print("Date Achieved:", achievement[1])
```

```python
# 3. Retrieve all habit names and their frequencies
cursor.execute("SELECT Habit_Name, Frequency FROM HABIT")
habits = cursor.fetchall()
print("\nHabits:")
for habit in habits:
    print("Habit Name:", habit[0])
    print("Frequency:", habit[1])
```

```python
# 4. Retrieve all networking event names and their dates
cursor.execute("SELECT Event_Name, Date FROM NETWORKING_EVENT")
networking_events = cursor.fetchall()
print("\nNetworking Events:")
for event in networking_events:
    print("Event Name:", event[0])
    print("Date:", event[1])
```

```python
# 5. Retrieve all personal goal names and their deadlines
cursor.execute("SELECT Goal_Name, Deadline FROM PERSONAL_GOAL")
personal_goals = cursor.fetchall()
print("\nPersonal Goals:")
for goal in personal_goals:
    print("Goal Name:", goal[0])
    print("Deadline:", goal[1])
```

```python
# 6. Retrieve all job application statuses
cursor.execute("SELECT DISTINCT Application_Status FROM JOB_APPLICATION")
application_statuses = cursor.fetchall()
print("\nJob Application Statuses:")
for status in application_statuses:
    print(status[0])
```

```python
# 7. Retrieve all mental health journal entries with their corresponding dates
cursor.execute("SELECT Journal_Entry, Date FROM MENTAL_HEALTH_ENTRY")
mental_health_entries = cursor.fetchall()
print("\nMental Health Journal Entries:")
for entry in mental_health_entries:
    print("Date:", entry[1])
    print("Journal Entry:", entry[0])
```

```python
# 8. Retrieve all professional types
cursor.execute("SELECT DISTINCT Professional_Type FROM PROFESSIONAL")
professional_types = cursor.fetchall()
print("\nProfessional Types:")
for prof_type in professional_types:
    print(prof_type[0])
```

```python
# Non simple Queries 
# 1. Retrieve the average score of all grades
cursor.execute("SELECT AVG(Score) FROM GRADE")
average_score = cursor.fetchone()[0]
print("\nAverage Score:", average_score)
```

```python

# 2. Retrieve the tasks assigned for courses along with their academic types
cursor.execute("SELECT c.Task_Name, a.Academic_Type FROM COURSE_ASSIGNMENT c INNER JOIN ACADEMIC a ON c.Academic_ID = a.Academic_ID")
course_assignments = cursor.fetchall()
print("\nCourse Assignments:")
for assignment in course_assignments:
    print("Task Name:", assignment[0])
    print("Academic Type:", assignment[1])
```

```python
# 3. Retrieve the total number of achievements for each user
cursor.execute("SELECT u.Name_Fname, u.Name_Lname, COUNT(a.Achievement_ID) AS Total_Achievements "
               "FROM USER u "
               "LEFT JOIN ACHIEVEMENT a ON u.User_ID = a.User_ID "
               "GROUP BY u.User_ID")
user_achievements = cursor.fetchall()
print("\nTotal Achievements per User:")
for user in user_achievements:
    print("User:", user[0], user[1])
    print("Total Achievements:", user[2])
```

```python
# 4. Retrieve the average difficulty level of tasks for each academic type
cursor.execute("SELECT a.Academic_Type, AVG(c.Difficulty) AS Avg_Difficulty "
               "FROM COURSE_ASSIGNMENT c "
               "INNER JOIN ACADEMIC a ON c.Academic_ID = a.Academic_ID "
               "GROUP BY a.Academic_Type")
academic_difficulty = cursor.fetchall()
print("\nAverage Difficulty Level per Academic Type:")
for academic_type in academic_difficulty:
    print("Academic Type:", academic_type[0])
    print("Average Difficulty:", academic_type[1])
```

```python
# 5. Retrieve the top 3 most frequently occurring habits
cursor.execute("SELECT Habit_Name, COUNT(*) AS Frequency "
               "FROM HABIT "
               "GROUP BY Habit_Name "
               "ORDER BY Frequency DESC "
               "LIMIT 3")
top_habits = cursor.fetchall()
print("\nTop 3 Most Frequently Occurring Habits:")
for habit in top_habits:
    print("Habit Name:", habit[0])
    print("Frequency:", habit[1])
```

```python
# 6. Retrieve the number of users who have made mental health entries on each date
cursor.execute("SELECT Date, COUNT(DISTINCT User_ID) AS Num_Users "
               "FROM MENTAL_HEALTH_ENTRY "
               "GROUP BY Date")
mental_health_users = cursor.fetchall()
print("\nNumber of Users Making Mental Health Entries per Date:")
for entry in mental_health_users:
    print("Date:", entry[0])
    print("Number of Users:", entry[1])
```

```python
# 7. Retrieve the total number of networking contacts for each professional type
cursor.execute("SELECT p.Professional_Type, COUNT(nc.Contact_ID) AS Total_Contacts "
               "FROM PROFESSIONAL p "
               "LEFT JOIN NETWORKING_CONTACT nc ON p.Professional_ID = nc.Professional_ID "
               "GROUP BY p.Professional_ID")
professional_contacts = cursor.fetchall()
print("\nTotal Networking Contacts per Professional Type:")
for prof_type in professional_contacts:
    print("Professional Type:", prof_type[0])
    print("Total Contacts:", prof_type[1])
```

```python
# 8. Retrieve the average score for each personal goal category
cursor.execute("SELECT p.Category, AVG(g.Score) AS Avg_Score "
               "FROM PERSONAL_GOAL p "
               "LEFT JOIN GRADE g ON p.Goal_ID = g.Academic_ID "
               "GROUP BY p.Category")
goal_scores = cursor.fetchall()
print("\nAverage Score for Each Personal Goal Category:")
for category in goal_scores:
    print("Category:", category[0])
    print("Average Score:", category[1])
```

```python
# 9. Retrieve the number of job applications for each company
cursor.execute("SELECT Company_Name, COUNT(*) AS Num_Applications "
               "FROM JOB_APPLICATION "
               "GROUP BY Company_Name")
company_applications = cursor.fetchall()
print("\nNumber of Job Applications per Company:")
for company in company_applications:
    print("Company Name:", company[0])
    print("Number of Applications:", company[1])
```

```python
# 10. Retrieve the average mood score for each user on their mental health entries
cursor.execute("SELECT u.Name_Fname, u.Name_Lname, AVG(mh.Mood) AS Avg_Mood "
               "FROM USER u "
               "INNER JOIN MENTAL_HEALTH_ENTRY mh ON u.User_ID = mh.User_ID "
               "GROUP BY u.User_ID")
user_mood_scores = cursor.fetchall()
print("\nAverage Mood Score per User:")
for user in user_mood_scores:
    print("User:", user[0], user[1])
    print("Average Mood Score:", user[2])
```

```python
#Query data for the plots 

# 1. Histogram: Distribution of scores for achievements
cursor.execute("SELECT Score FROM GRADE)
scores = cursor.fetchall()
scores = [score[0] for score in scores]

# Plotting

# 1. Histogram
plt.figure(figsize=(8, 6))
plt.hist(scores, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Scores for Achievements')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Bar Chart 
cursor.execute("SELECT p.Professional_Type, COUNT(nc.Contact_ID) AS Total_Contacts "
               "FROM PROFESSIONAL p "
               "LEFT JOIN NETWORKING_CONTACT nc ON p.Professional_ID = nc.Professional_ID "
               "GROUP BY p.Professional_ID")
professional_contacts_data = cursor.fetchall()
professional_types_bar = [data[0] for data in professional_contacts_data]
total_contacts = [data[1] for data in professional_contacts_data]

plt.figure(figsize=(10, 6))
plt.barh(professional_types_bar, total_contacts, color='orange')
plt.title('Total Number of Networking Contacts for Each Professional Type')
plt.ylabel('Professional Type')
plt.xlabel('Total Contacts')
plt.grid(axis='x')
plt.show()
```

```python
# Query data for box plot
cursor.execute("SELECT c.Difficulty, g.Score "
               "FROM GRADE g "
               "INNER JOIN COURSE_ASSIGNMENT c ON g.Task_ID = c.Task_ID")

# Fetch all rows
data = cursor.fetchall()

# Convert data to DataFrame
df = pd.DataFrame(data, columns=['Difficulty', 'Score'])

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Difficulty', y='Score')
plt.title('Distribution of Scores Achieved for Different Difficulty Levels in Course Assignments')
plt.xlabel('Difficulty Level')
plt.ylabel('Score Achieved')
plt.grid(True)
plt.show()
```

```python
# Query data for heatmap (for user 1)
cursor.execute("SELECT Industry, COUNT(Contact_ID) AS Total_Contacts "
               "FROM NETWORKING_CONTACT "
               "WHERE Professional_ID = 1 "
               "GROUP BY Industry")

# Fetch all rows
data = cursor.fetchall()
# Convert data to DataFrame
df = pd.DataFrame(data, columns=['Industry', 'Total_Contacts'])

# Pivot table for heatmap
pivot_table = df.pivot(index=None, columns='Industry', values='Total_Contacts').fillna(0)

# Plotting heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=0.5, linecolor='white', cbar=True)
plt.title('Networking Contacts Heatmap: Distribution among Industries')
plt.xlabel('Industry')
plt.ylabel('Frequency')
plt.show()
```
